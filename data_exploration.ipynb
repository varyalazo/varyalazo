```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
data_file = "/ceph/imaging1/davide/2p_data/429420_toms/20230210/neural_data.pickle"
filehandler = open(data_file, 'rb')
neural_data = pickle.load(filehandler)
filehandler.close()
```

```python
# look at dictionary keys
neural_data.keys()
```

```python
# put data in variables for easier use
traces = neural_data['traces']
spikes = neural_data['deconvolved']
positions = neural_data['positions']
```

```python
# visualize poisiton
plt.scatter(positions[:,1],positions[:,0])
```

```python
# visualize traces and spikes of a given neuron, in a given time slice
# the trace is z-scored: https://nl.wikipedia.org/wiki/Z-score

import matplotlib.pyplot as plt
from scipy.stats import zscore

neuron = 1
plt.plot(zscore(traces[neuron]))
plt.eventplot(spikes[neuron])
plt.xlim([0,1000])
plt.ylim([-1,5])
```
